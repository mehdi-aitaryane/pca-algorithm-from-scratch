```python
from dim_reduction import PCA
from datasets import make_classification, make_blobs, make_circles, print_shape
from plots import scatter2D, scatter3D
```

```python
X, y = make_classification(n_samples=1000, n_features=5, n_classes=2, balanced=True, noise=0.0, random_state=33)
```

```python
print_shape(X, y)
```

```python
title = "2D Original Make Classification Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X, y)
```

```python
title = "3D Original Make Classification Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
zlabel = "Feature 3"
scatter3D(title, xlabel, ylabel, zlabel, X, y)
```

```python
pca = PCA(n_components=2)
X_trans = pca.fit_transform(X)
```

```python
print_shape(X_trans, y)
```

```python
title = "2D PCA Make Classification Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X_trans, y)
```

```python
pca = PCA(n_components=3)
X_trans = pca.fit_transform(X)
```

```python
title = "3D PCA Make Classification Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
zlabel = "Feature 3"
scatter3D(title, xlabel, ylabel, zlabel, X_trans, y)
```
